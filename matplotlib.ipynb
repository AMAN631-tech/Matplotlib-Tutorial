# Introduction to Matplotlib

Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats. Developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding. If you want to convert your boring data into interactive plots and graphs, Matplotlib is the tool for you.

# Example of a Plot in Matplotlib:
Let's create a simple line plot using Matplotlib, showcasing the ease with which you can visualize data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1️⃣ Line Plot
What is it?

A Line Plot is a basic graph that shows data points connected by lines.
It is used to show trends, changes, or relationships between two variables.



```python
# Single Line Plot

x = [1,2,3,4,5]
y = [1,3,6,5,9]

plt.plot(x,y)
plt.show()
```

```python
# Multiple Line Plot

x = [1,2,3,4,5]
y = [1,3,6,5,9]

plt.plot(x,y,color ='red',linestyle='dotted',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
z = [2,3,6,8,1]
y = [1,3,6,5,9]

plt.plot(x,y,color ='red',label='X',linestyle='dotted',marker='o',linewidth=2,markersize=8)
plt.plot(x,z,color ='blue',label='Z',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

# 2️⃣ Subplot

What is it?

A subplot is a grid of multiple plots in a single figure.
You use subplots when you want to show multiple graphs side by side.

```python

x = [1,2,3,4,5]
z = [2,3,6,8,1]
y = [1,3,6,5,9]
w = [i*2 for i in z]
plt.figure(figsize=(10,8))

# figsize = (width, height)
# 10 = Width (horizontal size)

# 8 = Height (vertical size)

plt.subplot(2,2,1)
plt.plot(x,y,color ='red')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(x,z,color ='green')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(z,y,color ='blue')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(z,w,color ='black')
plt.title('Plot 4')

plt.show()
```

#  What is a Bar Plot?
A Bar Plot (or Bar Chart) is a graph that uses rectangular bars to show the size, quantity, or frequency of categories.

Bars can be vertical or horizontal.

The length/height of the bar = value of the category.

# Why use a Bar Plot?
Bar plots are used to:

| **Purpose**               | **Example**                       |
| ------------------------- | --------------------------------- |
| Compare **categories**    | Sales by Product                  |
| Show **counts/frequency** | Number of students in each grade  |
| Visualize **group data**  | Population of different countries |



```python
categories = ['A','B','C','D']
values = [10,20,30,40]
plt.bar(categories,values,color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

# What is a Histogram?
A Histogram is a special kind of bar graph that shows the distribution of numerical data.
It tells you how many values fall into specific ranges (called "bins").

# Difference Between Bar Plot and Histogram:

| **Bar Plot**                   | **Histogram**                      |
| ------------------------------ | ---------------------------------- |
| Used for **categories**        | Used for **numeric ranges**        |
| Bars can be **separate**       | Bars are usually **touching**      |
| Shows **counts of categories** | Shows **frequency of data ranges** |


# Why use a Histogram?

To answer questions like:

.How are the values spread?

.Are the data values mostly small, large, or balanced?

.Is the data normal, skewed, or has outliers?



```python
data = [50,65,32,15,99,47,67,59,83,71]

plt.figure(figsize=(7,5))
plt.hist(data,bins=5,color='orange',edgecolor='white')
```

| **Axis**   | **What it shows**                |
| ---------- | -------------------------------- |
| **X-axis** | **Value Ranges (Bins)**          |
| **Y-axis** | **Number of values (Frequency)** |


# What is a Scatter Plot?
A Scatter Plot is a graph where each point represents two variables:

One on the x-axis

One on the y-axis

It shows how two numeric variables are related.

# Why use a Scatter Plot?
| **Use**                   | **Purpose**                                         |
| ------------------------- | --------------------------------------------------- |
| **Find relationships**    | See if one variable affects the other               |
| **Detect patterns**       | Check for **trends**, **clusters**, or **outliers** |
| **Visualize correlation** | Positive, negative, or no relation                  |

# Real-Life Example:

| **x-axis (Independent Variable)** | **y-axis (Dependent Variable)** |
| --------------------------------- | ------------------------------- |
| Study hours                       | Exam scores                     |
| Temperature                       | Ice cream sales                 |
| Advertising cost                  | Sales revenue                   |


```python
study_hours = [1, 2, 3, 4, 5, 6]
exam_scores = [50, 55, 65, 70, 75, 85]

plt.scatter(study_hours,exam_scores,color='red',marker='o')
# Add labels and title
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')

plt.grid(True)
plt.show()
```

# Types of corelation in Scatter plot.

| **Type**                 | **How It Looks**                | **Meaning**                     |
| ------------------------ | ------------------------------- | ------------------------------- |
| **Positive Correlation** | Dots go **upward** together     | As **X increases, Y increases** |
| **Negative Correlation** | Dots go **downward** together   | As **X increases, Y decreases** |
| **No Correlation**       | Dots are **scattered randomly** | No clear relationship           |


# Pie Chart

# What is a Pie Chart?
A Pie Chart is a circular graph divided into slices (sections), where:

Each slice = a category’s contribution to the total.

The size of the slice = percentage or proportion of that category.

# Why use a Pie Chart?

| **Use Case**                        | **Example**                          |
| ----------------------------------- | ------------------------------------ |
| Show **part-to-whole relationship** | Market share of companies            |
| Display **percentages**             | Population by religion               |
| Visualize **category contribution** | Budget spending (food, rent, travel) |


```python
labels = ['Food', 'Rent', 'Bills', 'Savings']
sizes = [40, 30, 20, 10]

plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
df = pd.read_csv('sales_data1.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head()
```

```python
# plotting the line plot..
plt.figure(figsize=(8,6))
plt.plot(df['Value'],df['Sales'],color='orange',linestyle='--',marker='o')
plt.show()
```

```python
df.head()
```

```python
# bar plot

plt.figure(figsize=(8,6))
plt.bar(df['Category'],df['Sales'],color='black')
plt.show()
```

```python
# plotting Histogram..

plt.figure(figsize=(8,6))
plt.hist(df['Sales'],bins=6,color='black',edgecolor='white')
```

```python
# plotting the scatter plot..
plt.figure(figsize=(8,6))
plt.scatter(df['Value'],df['Sales'],color='orange',marker='o')
plt.show()
```

# "No Correlation Scatter Plot"
Why?
The points are scattered everywhere

No visible upward or downward trend

No relationship between X (Value) and Y (Sales)



```python
# plotting pie chart..

val=df['Category'].value_counts()
val
```

```python
plt.figure(figsize=(8,6))
plt.pie(val,autopct='%1.1f%%',labels=val.index,startangle=90)
plt.show()
```

```python
# Group by Category and sum Sales
sales_by_category = df.groupby('Category')['Sales'].sum()

# Create pie chart
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', startangle=90)

plt.title('Sales Distribution by Category')
plt.show()
```
